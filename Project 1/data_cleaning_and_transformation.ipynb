```python
# Import Packages
import pandas as pd
```

```python
# Reading the CSV
df = pd.read_csv('dataset/Cleaned_DS_Jobs.csv')

# Show the first 5 rows of df
df.head()
```

1. Make the 'Salary Estimate' column into floats

```python
def calculate_average_from_range(range) -> float:
    #range example: "100-120"
    lower, upper = map(int, range.split("-"))
    average = (upper + lower) / 2
    return average

df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: calculate_average_from_range(x))
df.head()
```

2. Delete everthing after ',' in column "Headquarters" and "Location"

```python
df['Headquarters'] = df['Headquarters'].apply(lambda x: x.split(',')[0])
df['Location'] = df['Location'].apply(lambda x: x.split(',')[0])
df[['Headquarters', 'Location']]
```

3. Create new column with average 'Size'

```python
def calculate_average_size(x):
    parts = x.split(" ")
    if len(parts) > 3:
        a = parts[0]
        b = parts[2]
        a = int(a)
        b = int(b)
        average = (a+b) / 2
        return average
    else:
        return x
    
    
df['Average Size'] = df['Size'].apply(lambda x: calculate_average_size(x))
df[['Size', 'Average Size']].head()
```
